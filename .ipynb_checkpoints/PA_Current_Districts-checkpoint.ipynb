# Analysis of Current Pennsylvania Congressional Districts

In this notebook, we will compile and analyze geospatial data for the 2022 Congressional Districts of Pennsylvania.

```python
import geopandas as gpd
```

### Congressional Districts

Below is the process of importing the shapefiles of the Congressional Districts for all 50 states. This data is accredited to the U.S. Census.

```python
df = gpd.read_file('data/US_con_dists/cb_2022_us_cd118_500k.shp')
```

```python
# Extracting data from Pennsylvania
df['area'] = df.area
df_pa = df[df['STATEFP'] == '42']
df_pa.head()
```

```python
# Displaying the 13 districts of PA
df_pa.plot('NAMELSAD', cmap='nipy_spectral')
```

### Voting Districts data for PA

These shapefiles are accredited to the Metric Geometry and Gerrymandering Group (MGGG).

```python
pa_vtds = gpd.read_file('data/PA_pop/PA_pop.shp')
```

```python
pa_vtds.head()
```

```python
# Get rid of Lake Erie census block
pa_vtds['AREA'] = pa_vtds.area
pa_vtds = pa_vtds[pa_vtds['NAME'] != 'LAKE ERIE']
```

```python
# Display all Pennsylvania voting districts
pa_vtds.plot('NAMELSAD20')
```

### Analysis Dataset

```python
# Merging spatial data: set the CRS of the congressional districts to that
# of the voting districts dataframe so they can be merged
df_pa = df_pa.to_crs(epsg=6562)
blocks_in_districts = gpd.sjoin(pa_vtds, df_pa)
blocks_in_districts.head()
```

```python
# Cleaning 
pa_cong_dists = blocks_in_districts.drop([
    'AFFGEOID', 'GEOID', 'STATEFP_left', 'STATEFP_right', 'COUNTYFP', 'VTDST20',
    'COUNTYFP20', 'VTDI20', 'index_right', 'CDSESSN'], axis=1)
pa_cong_dists = pa_cong_dists.rename(columns={'G18GOVRWAG': 'rep_votes_governor18',
                                              'NAME': 'district_name',
                                              'VTDST': 'district_code',
                                              'G18USSRBAR': 'rep_votes_senate18',
                                              'G18USSDCAS': 'dem_votes_senate18',
                                              'G18GOVDWOL':'dem_votes_governor18',
                                              'G16PRERTRU':'rep_votes_president16',
                                              'G16PREDCLI':'dem_votes_president16',
                                              'AREA': 'area_vtds',
                                              'area': 'area_cong_dists',
                                              'GEOID20': 'geographic_ID'})
```

```python
pa_cong_dists.plot('NAMELSAD')
```

## Scoring

### Efficency Gap (EG):
a district is "fair" if two parties waste an equal number of votes (EG = 0). Any votes for a canidate over 50% are considered wasted, and all losing votes are wasted

rep_votes_governor18: Number of votes received by the Republican candidate for governor in the 2018 general election in this voting district.

dem_votes_governor18: Number of votes received by the Democratic candidate for governor in the 2018 general election in this voting district.

```python
def calcEG(data):
# calcEG returns the efficency gap for the given district and adds to the global totals for the state EG
    global total_wasted_A_votes
    global total_wasted_B_votes
    global total_votes_PA
    
    # A should be republican
    A_votes = data['rep_votes_governor18'].sum()
    # B should be democrat
    B_votes = data['dem_votes_governor18'].sum()
    total_votes = A_votes + B_votes
    votes_to_win = int(total_votes/2) + 1
    winner = max(A_votes,B_votes)
    loser = min(A_votes,B_votes)

    # winning party = wasted over 50% 
    wasted_over_50 = winner - votes_to_win
    
    # losing party = all wasted
    if winner == A_votes:
        total_wasted_A_votes += wasted_over_50
        total_wasted_B_votes += B_votes
    else:
        total_wasted_B_votes += wasted_over_50
        total_wasted_A_votes += A_votes
        
    total_votes_PA += total_votes
    
    if total_votes != 0:
        return(wasted_over_50-loser)/total_votes
    else:
        return(0)
```

```python
all_districts = pa_cong_dists.groupby('NAMELSAD')
total_wasted_A_votes,total_wasted_B_votes,total_votes_PA = 0,0,0
EG_DATA = []
for district, data in all_districts:
    EG_DATA.append((calcEG(data),district))
```

```python
print(EG_DATA)
```

```python
dem_favored = 0
rep_favored = 0
equal = 0

for d in EG_DATA:
    # dem favored if EG less than 0
    if d[0] < 0:
        dem_favored += 1
    # rep favored if EG greater than 0
    elif d[0] > 0:
        rep_favored += 1   
    elif d[0] == 0:
        equal += 1
        
EG = (total_wasted_A_votes - total_wasted_B_votes)/total_votes_PA

print('number dem favored districts: ',dem_favored)
print('number rep favored districts: ',rep_favored)
print('number equal districts: ',equal)
print('PA EG: ',EG)
```
